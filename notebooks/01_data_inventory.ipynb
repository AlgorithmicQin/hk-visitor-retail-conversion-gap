Place manually downloaded monthly files in `data/raw/`, then run this notebook to inspect file names, columns, and sample rows before mapping.

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().resolve()
if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))

RAW_DIR = PROJECT_ROOT / "data" / "raw"
raw_files = sorted([p for p in RAW_DIR.glob("*") if p.is_file()])
raw_files
```

```python
from src.data_loader import read_tabular_file

for path in raw_files:
    print(f"\n=== {path.name} ===")
    try:
        sample = read_tabular_file(path, nrows=5)
        print(list(sample.columns))
        display(sample)
    except Exception as exc:
        print(f"Could not preview {path.name}: {exc}")
```
