# 03 Conversion Gap Pilot

This notebook calculates visitor-retail and visitor-hotel recovery gaps from the indexed pilot panel. A positive gap means the outcome index is above the visitor arrivals index; a negative gap means it is below visitor recovery.

```python
from pathlib import Path
import sys
import pandas as pd

PROJECT_ROOT = Path.cwd().resolve()
if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))

TABLE_DIR = PROJECT_ROOT / "outputs" / "tables"
FIGURE_DIR = PROJECT_ROOT / "outputs" / "figures"
TABLE_DIR.mkdir(parents=True, exist_ok=True)
FIGURE_DIR.mkdir(parents=True, exist_ok=True)
```

```python
from src.index_builder import calculate_conversion_gaps, select_gap_columns
from src.plotting import plot_conversion_gaps

indexed = pd.read_csv(TABLE_DIR / "recovery_indices.csv", parse_dates=["month"])
gaps = calculate_conversion_gaps(indexed)
gaps.to_csv(TABLE_DIR / "conversion_gaps.csv", index=False)
gaps.head()
```

```python
gap_columns = select_gap_columns(gaps)
fig, ax = plot_conversion_gaps(
    gaps,
    gap_columns,
    output_path=FIGURE_DIR / "conversion_gaps.png",
)
```
